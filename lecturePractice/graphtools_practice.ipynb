# `matplotlib`
* 다양한 그래프 지원
* `pandas` 지원

```python
import matplotlib.pyplot as plt
import numpy as np

X = range(100)
Y = range(100)

plt.plot(X, Y)
```

```python
X_1 = range(100)
Y_1 = [np.cos(value) for value in X_1]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X_2]

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.show()
```

```python
import random

plt.plot([random.random() for _ in range(100)], range(100))
```

```python
X_1 = range(100)
Y_1 = [np.cos(value) for value in X_1]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X_2]

fig = plt.figure()
fig.set_size_inches(10, 5)
ax_1 = fig.add_subplot(1, 2, 1)
ax_2 =fig.add_subplot(1, 2, 2)
ax_1.plot(X_1, Y_1)
ax_2.plot(X_2, Y_2)
```

```python
data_1 = np.random.rand(512, 2)
data_2 = np.random.rand(512, 2)

plt.scatter(data_1[:, 0], data_1[:, 1], c='b', marker='x')
plt.scatter(data_2[:, 0], data_2[:, 1], c='r', marker='^')
```

```python
data = [
    [5, 25, 50, 20],
    [4, 23, 51, 17],
    [6, 22, 52, 19]
]

x=np.arange(4)
plt.bar(x + 0, data[0], color='b', width=0.25)
plt.bar(x + 0.25, data[1], color='g', width=0.25)
plt.bar(x + 0.5, data[2], color='r', width=0.25)
plt.xticks(x+0.25, ('A','B','C','D'))
```

```python
col_list = ['b', 'g', 'r']
d_label = ['A', 'B', 'C']
x = np.arange(np.array(data).shape[1])
for i in range(3):
    plt.bar(
        x,
        data[i],
        bottom=np.sum(data[:i], axis=0),
        color=col_list[i],
        label=d_label[i]
    )
plt.legend()
plt.show()
```

# `seaborn`
* statistical data visualization

```python
import seaborn as sns

sns.set(style='darkgrid')
```

```python
tips = sns.load_dataset('tips')
fmri = sns.load_dataset('fmri')
```

```python
sns.set_style('whitegrid')
sns.lineplot(x='timepoint', y='signal', data=fmri)

# 주요 데이터를 연결하고 데이터 분포를 흐린 색으로 표현
```

```python
sns.lineplot(x='timepoint', y='signal', hue="event", data=fmri)

# event에 따라 categorize 후 그래프화
```
