## Image Translation GANs

### Pix2Pix
![](0311_imgs/pix2pix.png)
* 하나의 이미지를 대응하는 다른 도메인의 이미지로 변환하는 Task
* Loss 계산 시 L1 Loss와 GAN Loss를 모두 고려
    * L1 Loss: x를 기반으로 생성한 이미지가 y에 가깝도록 & GAN의 안정화를 위해 사용
    * GAN Loss: 실제에 가까운 형태의 이미지를 생성하기 위해 사용
    
    
$$
G^* = \text{arg}\underset{G}{\text{min}}\underset{D}{\text{max}}\mathcal{L}_{cGAN}(G, D) + \lambda\mathcal{L}_{L1}(G) \\
\mathcal{L}_{cGAN}(G, D) = \mathbb{E}_{x, y}\left[\text{log}D(x, y)\right] + \mathbb{E}_{x, z}\left[\text{log}(1 - D(x, G(x, z)))\right] \\
\mathcal{L}_{L1}(G) = \mathbb{E}_{x, y,z}\left[||y-G(x, z)||_1\right]
$$

* Loss 비교
![](0311_imgs/pix2pix_comp.png)


### CycleGAN
![](0311_imgs/cyclegan.png)
* Pix2Pix를 학습하려면 일대일로 대응되는 이미지 pair가 필요
* CycleGAN: 일대일 대응 없이 각 Domain Image Set 만으로도 결과를 도출 가능

![](0311_imgs/cyclegan_architecture.png)
* CycleGAN Loss = GAN Loss + Cycle-Consistency Loss
    * GAN Loss: 각 도메인으로 이미지를 변환하는 부분에 대해 학습
    * Cycle-Consistency Loss: Generator가 입력된 이미지의 특성을 제대로 활용하는지 판단
![](0311_imgs/cyclegan_loss.png)
![](0311_imgs/cycle_consistency_loss.png)

### Perceptual Loss
* 기존의 GAN은 학습하기 매우 어려움 -> GAN 없이 좋은 결과를 내기 위한 Loss
* Perceptual Loss의 개념: Pre-trained Classifier의 레이어를 이용해 이미지를 변환

![](0311_imgs/perception_loss.png)
* Perceptual Loss를 이용한 Style Transfer
* Image Transform Net + Loss Net의 형태
* 2개의 Perceptual Loss 사용
    * $\phi_j$: j번째 Conv Layer의 Feature Map
    * Style Reconstruction Loss
